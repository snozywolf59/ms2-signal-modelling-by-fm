```python
from datasets import load_dataset
```

```python
data = load_dataset(path="../data/Pep2Prob/data_split/train_test_split_set_1")
# test = load_dataset(path="../data/Pep2Prob/data/", split="test")
```

```python
# đọc 100 row đầu tiên trong tập train
print(data["train"][0])
```

```python
import matplotlib.pyplot as plt

# Lấy peptide_length từ tập train
peptide_lengths = [item["peptide_length"] for item in data["train"]]

# Vẽ histogram
plt.figure()
plt.hist(peptide_lengths, bins=30)
plt.xlabel("Peptide length")
plt.ylabel("Count")
plt.title("Distribution of Peptide Lengths (Train set)")
plt.show()
```
