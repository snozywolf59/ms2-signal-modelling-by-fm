```python
import h5py
import numpy as np
```

```python
train_path = r"E:\Dai hoc\2526I\dacn\flow-matching\data\traintest_hcd.hdf5"
test_path = r"E:\Dai hoc\2526I\dacn\flow-matching\data\holdout_hcd.hdf5"
```

```python

with h5py.File(train_path, "r") as f:
    print("Keys:", list(f.keys()))

    train_seqs = f["sequence_integer"][:]
    train_charges = f["precursor_charge_onehot"][:]

with h5py.File(test_path, "r") as f:
    print("Keys:", list(f.keys()))

    test_seqs = f["sequence_integer"][:]    
    test_charges = f["precursor_charge_onehot"][:]
    
```

```python
train_charges = np.argmax(train_charges, axis=1) + 1
test_charges = np.argmax(test_charges, axis=1) + 1
```

### Check for duplicate precursors

```python
train_set = {tuple(seq): i for i, seq in enumerate(train_seqs)}
```

```python
test_charges[0:9]
```

```python
duplicate = []
for i, seq in enumerate(test_seqs):
    seq_t = tuple(seq)
    if seq_t in train_set.keys():
        j = train_set[seq_t]
        if train_charges[j] == test_charges[i]:
            duplicate.append((i, j))
```
