```python
from collections import defaultdict

from nist.process_msp import read_msp
```

```python
msp_path = r"E:\Dai hoc\2526I\dacn\flow-matching\nist\human_hcd_tryp_best.msp"
cnt_pep = defaultdict(int)
for spectrum in read_msp(msp_path):
    pep_name = spectrum["Name"].split("_")[0]
    cnt_pep[pep_name] += 1
    
```
