```python
import sys
sys.path.append(r"E:\Dai hoc\2526I\dacn\flow-matching")
import h5py
from collections import defaultdict, Counter
import numpy as np
```

```python
file_path = r"E:\Dai hoc\2526I\dacn\flow-matching\data\traintest_hcd.hdf5"
```

```python
with h5py.File(file_path, "r") as f:
    print("Keys:", list(f.keys()))
    seqs = f["sequence_integer"][:]
```

## Most frequent pep

```python
seqs_tuple = [tuple(seq) for seq in seqs]
counter = Counter(seqs_tuple)
most_freq_peptide, freq = counter.most_common(1)[0]
print("Most frequent peptide:", most_freq_peptide)
print("Frequency:", freq)
```

## Take raw_intensities of above pep

```python
with h5py.File(file_path, "r") as f:
    intensities = f["intensities_raw"][:]
    
ids = [i for i, seq in enumerate(seqs_tuple) if seq == most_freq_peptide]

print("Number of spectra:", len(ids))
selected_intensities = intensities[ids]

print("First spectrum intensities:", selected_intensities[0])
```
